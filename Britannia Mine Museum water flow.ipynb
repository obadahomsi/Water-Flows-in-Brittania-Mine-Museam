```python
grd_water = 'BA-10037'
acid_water = 'BA-10182'
discharge = 'BA-10078'


def segregate_to_months(dataframe,stamp):
    months = ['03','04','05','06','07','08','09','10','11','12','01','02','03']
    temp_storage = [[],[],[],[],[],[],[],[],[],[],[],[],[]]
    for i in range(len(dataframe['Timestamp'])):
        
        if (dataframe['Timestamp'][i][5] +dataframe['Timestamp'][i][6]==months[0]) and dataframe['Timestamp'][i][3]=='8': 
            temp_storage[0].append(dataframe[stamp][i])  
        
        elif (dataframe['Timestamp'][i][5] +dataframe['Timestamp'][i][6]==months[1]) and dataframe['Timestamp'][i][3]=='8': 
            temp_storage[1].append(dataframe[stamp][i])

        elif (dataframe['Timestamp'][i][5] +dataframe['Timestamp'][i][6]==months[2]) and dataframe['Timestamp'][i][3]=='8': 
            temp_storage[2].append(dataframe[stamp][i])

        elif dataframe['Timestamp'][i][5] +dataframe['Timestamp'][i][6]==months[3]: 
            temp_storage[3].append(dataframe[stamp][i])

        elif dataframe['Timestamp'][i][5] +dataframe['Timestamp'][i][6]==months[4]: 
            temp_storage[4].append(dataframe[stamp][i])

        elif dataframe['Timestamp'][i][5] +dataframe['Timestamp'][i][6]==months[5]: 
            temp_storage[5].append(dataframe[stamp][i])

        elif dataframe['Timestamp'][i][5] +dataframe['Timestamp'][i][6]==months[6]: 
            temp_storage[6].append(dataframe[stamp][i])

        elif dataframe['Timestamp'][i][5] +dataframe['Timestamp'][i][6]==months[7]: 
            temp_storage[7].append(dataframe[stamp][i])

        elif dataframe['Timestamp'][i][5] +dataframe['Timestamp'][i][6]==months[8]: 
            temp_storage[8].append(dataframe[stamp][i])

        elif dataframe['Timestamp'][i][5] +dataframe['Timestamp'][i][6]==months[9]: 
            temp_storage[9].append(dataframe[stamp][i])

        elif dataframe['Timestamp'][i][5] +dataframe['Timestamp'][i][6]==months[10]: 
            temp_storage[10].append(dataframe[stamp][i])

        elif dataframe['Timestamp'][i][5] +dataframe['Timestamp'][i][6]==months[11]: 
            temp_storage[11].append(dataframe[stamp][i])
            
        elif dataframe['Timestamp'][i][5] +dataframe['Timestamp'][i][6]==months[12]: 
            temp_storage[12].append(dataframe[stamp][i])

    return temp_storage 
                            
                    
        
    
```

```python
gw = segregate_to_months(groundwater_ph,grd_water)
aw = segregate_to_months(acidic_water_ph,acid_water)
dw = segregate_to_months(discharge_flow,discharge)
```
